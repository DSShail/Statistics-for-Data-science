### d. Cumulative Frequency
#### The frequency is the number of times an event occurs within a given scenario. Cumulative frequency is defined as the running total of frequencies. It is the sum of all the previous frequencies up to the current point. It is easily understandable through a Cumulative Frequency Table.

```python
a=df['Pclass'].value_counts()
```

```python
#calculate the percentage of each class
b=(a/a.sum()*100).values
```

```python
b
```

```python
c=[]
cf=0
for i in b:
    cf+=i
    c.append(cf)
```

```python
plt.plot(c,marker='d')
```

```python
iris_df['Species'].value_counts()
```

```python
iris_setosa=iris_df[iris_df['Species']=='Iris-setosa']
iris_versicolor=iris_df[iris_df['Species']=='Iris-versicolor']
iris_virginica=iris_df[iris_df['Species']=='Iris-virginica']
```

## Mean


```python
#sepal width

print(np.mean(iris_setosa['SW']))
print(np.mean(iris_versicolor['SW']))
print(np.mean(iris_virginica['SW']))
```

```python
#sepal length

print(np.mean(iris_setosa['SL']))
print(np.mean(iris_versicolor['SL']))
print(np.mean(iris_virginica['SL']))
```

```python
#petal length


print(np.mean(iris_setosa['PL']))
print(np.mean(iris_versicolor['PL']))
print(np.mean(iris_virginica['PL']))
```

```python
#petal width


print(np.mean(iris_setosa['PW']))
print(np.mean(iris_versicolor['PW']))
print(np.mean(iris_virginica['PW']))
```

```python
plt.hist(iris_setosa['PL'])
```

### Problem with Outliers

```python
np.mean(np.append(iris_setosa['PL'],50))
```

##### just adding a number shifs the mean from 1.4 to 2.4- this is the disadvantage of outliers and the solution is median

## Median
- sort the data in ascending order and the middle element is your median



```python
#petal length
print(np.median(iris_setosa['PL']))
print(np.median(iris_versicolor['PL']))
print(np.median(iris_virginica['PL']))
```

### The good thing about the median is that it doesn't depend on outlier

```python
np.median(np.append(iris_setosa['PL'],50))
```

##### addind 50 to iris_setosa didn't impact the median
